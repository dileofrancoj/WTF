```python
import numpy as np
import matplotlib.pyplot as plt

# Constante de Schwarzschild (podés ajustar a gusto)
rs = 1

# Valores de r para los que graficaremos las hipérbolas
r_values = [0.5, 0.9, 1.1, 2, 3, 5]

# Crear figura
plt.figure(figsize=(8, 8))

# Graficar hipérbolas para distintos valores de r
for r in r_values:
    factor = np.abs(r / rs - 1) * np.exp(r / rs)
    c = (rs**2 / 4) * factor

    # Dominio u para graficar (simétrico)
    u = np.linspace(-10, 10, 1000)
    v_squared = u**2 - c

    # Solo tomamos valores reales de v
    mask = v_squared >= 0
    u = u[mask]
    v_pos = np.sqrt(v_squared[mask])
    v_neg = -v_pos

    plt.plot(u, v_pos, label=f"r = {r}")
    plt.plot(u, v_neg)

# Horizonte: r = r_s => u^2 - v^2 = 0, o sea, u = ±v (las asíntotas)
x = np.linspace(-10, 10, 1000)
plt.plot(x, x, 'k--', label='Horizonte (u = v)')
plt.plot(x, -x, 'k--')

plt.xlabel("u")
plt.ylabel("v")
plt.title("Líneas de r = cte en coordenadas de Kruskal-Szekeres")
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

rs = 1
c = 1
t_values = [-10, -5, -2, 0, 2, 5, 10]
u = np.linspace(-10, 10, 1000)

plt.figure(figsize=(8, 8))

for t in t_values:
    slope = np.tanh(c * t / (2 * rs))
    v = slope * u
    plt.plot(u, v, label=f"t = {t}")

x = np.linspace(-10, 10, 1000)
plt.plot(x, x, 'k--', label='Horizonte u=v')
plt.plot(x, -x, 'k--')

plt.xlabel("u")
plt.ylabel("v")
plt.title("Líneas de $t = \\text{cte}$ en coordenadas de Kruskal-Szekeres")
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.tight_layout()
plt.savefig("kruskal_t_const.png", dpi=300)
plt.show()
```
