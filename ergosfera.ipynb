```python
import numpy as np
import matplotlib.pyplot as plt

# Constantes naturales (pods cambiarlas si quers que se vean las unidades reales)
G = 1
c = 1
M = 1  # masa del agujero negro
a = 0.2  # parmetro de rotacin (debe cumplir a <= GM/c)

# Definimos theta desde 0 a pi (corte meridional)
theta = np.linspace(0, np.pi, 1000)

# Horizonte de eventos (es constante para todas las direcciones)
r_plus = (G*M)/c**2 + np.sqrt((G*M/c**2)**2 - (a/c)**2)

# Radio de la ergosfera en funcin de theta
r_erg = (G*M)/c**2 + np.sqrt((G*M/c**2)**2 - (a/c)**2 * np.cos(theta)**2)

# Convertimos a coordenadas cartesianas para el plot
x_erg = r_erg * np.sin(theta)
y_erg = r_erg * np.cos(theta)

x_hor = r_plus * np.sin(theta)
y_hor = r_plus * np.cos(theta)

# Creamos la figura
plt.figure(figsize=(6, 6))
plt.plot(x_erg, y_erg, label='Ergosfera', color='orange')
plt.plot(-x_erg, y_erg, color='orange')  # simetra
plt.plot(x_hor, y_hor, label='Horizonte de eventos', color='black')
plt.plot(-x_hor, y_hor, color='black')

plt.title("Corte Meridional del Agujero Negro de Kerr")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constantes naturales
G = 1
c = 1
M = 1
a = 0.8  # debe cumplir a <= GM/c

# Coordenadas angulares
theta = np.linspace(0, np.pi, 200)
phi = np.linspace(0, 2*np.pi, 200)
theta, phi = np.meshgrid(theta, phi)

# Funciones de r para horizonte y ergosfera
r_plus = (G*M)/c**2 + np.sqrt((G*M/c**2)**2 - (a/c)**2)
r_erg = (G*M)/c**2 + np.sqrt((G*M/c**2)**2 - (a/c)**2 * np.cos(theta)**2)

# Conversin a coordenadas cartesianas
# Horizonte
x_h = r_plus * np.sin(theta) * np.cos(phi)
y_h = r_plus * np.sin(theta) * np.sin(phi)
z_h = r_plus * np.cos(theta)

# Ergosfera
x_e = r_erg * np.sin(theta) * np.cos(phi)
y_e = r_erg * np.sin(theta) * np.sin(phi)
z_e = r_erg * np.cos(theta)

# Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Superficie del horizonte
ax.plot_surface(x_h, y_h, z_h, color='black', alpha=0.8, label='Horizonte de eventos')
# Superficie de la ergosfera
ax.plot_surface(x_e, y_e, z_e, color='orange', alpha=0.5, label='Ergosfera')

# Esttica
ax.set_title("Agujero Negro de Kerr: Horizonte de Eventos y Ergosfera", fontsize=14)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_box_aspect([1, 1, 1])  # Aspecto igual
ax.grid(True)

# Leyenda manual
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='black', edgecolor='k', label='Horizonte de eventos'),
    Patch(facecolor='orange', edgecolor='k', label='Ergosfera')
]
ax.legend(handles=legend_elements, loc='upper right')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ngulo polar
theta = np.linspace(0, np.pi, 500)

# Parmetros de rotacin para las ergosferas
a_values = [0.3, 0.6, 0.9]

# Horizonte de eventos fijo para a = 0.8
a_hor = 0.8
r_plus = 1 + np.sqrt(1 - a_hor**2)
x_hor = r_plus * np.sin(theta)
y_hor = r_plus * np.cos(theta)

# Crear la figura
plt.figure(figsize=(8, 6))

# Dibujar las ergosferas para distintos a
for a in a_values:
    r_erg = 1 + np.sqrt(1 - a**2 * np.cos(theta)**2)
    x_erg = r_erg * np.sin(theta)
    y_erg = r_erg * np.cos(theta)
    plt.plot(x_erg, y_erg, label=f'Ergosfera (a={a})', linewidth=2)

# Dibujar el horizonte de eventos fijo
plt.plot(x_hor, y_hor, 'k--', label='Horizonte (a=0.8)', linewidth=2)

# Esttica
plt.title("Comparacin de ergosferas para distintos $a$ con un horizonte fijo (a = 0.8)")
plt.xlabel("x = r sin()")
plt.ylabel("y = r cos()")
plt.gca().set_aspect('equal')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
