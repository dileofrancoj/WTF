```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# Parámetros 
g = 9.81
L = 1.0 
theta_0 = 0.2
omega_0 = 0.0
t_max = 10  # Tiempo total de simulación (s)
```

```python
def pendulum(t, y):
    theta, omega = y
    dydt = [omega, -(g / L) * np.sin(theta)]  # [dot(theta), dot(omega)]
    return dydt

t_eval = np.linspace(0, t_max, 1000)

sol = solve_ivp(pendulum, [0, t_max], [theta_0, omega_0], t_eval=t_eval, method='RK45')


time = sol.t  # Tiempo
theta = sol.y[0]  # Posición angular
omega = sol.y[1]  # Velocidad angular
```

```python
plt.figure(figsize=(10, 6))

# Posición angular
plt.subplot(2, 1, 1)
plt.plot(time, theta, label="Posición angular (θ)", color='blue')
plt.xlabel("Tiempo (s)")
plt.ylabel("Posición angular (rad)")
plt.title("Posición y Velocidad Angular del Péndulo Simple")
plt.grid()
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(time, omega, label="Velocidad angular (ω)", color='green')
plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad angular (rad/s)")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
