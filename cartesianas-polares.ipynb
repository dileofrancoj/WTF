```python
import numpy as np
import matplotlib.pyplot as plt

# Crear una malla en coordenadas polares
r = np.linspace(0, 5, 25)
theta = np.linspace(0, 2*np.pi, 50)
R, Theta = np.meshgrid(r, theta)

# Transformacin a coordenadas cartesianas
X = R * np.cos(Theta)
Y = R * np.sin(Theta)

# Crear la figura con dos subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6), subplot_kw={'aspect': 'equal'})

# Subplot 1: Grilla cartesiana
x_vals = np.linspace(-5, 5, 20)
y_vals = np.linspace(-5, 5, 20)
X_grid, Y_grid = np.meshgrid(x_vals, y_vals)

axs[0].plot(X_grid, Y_grid, color='gray', linewidth=0.5)
axs[0].plot(X_grid.T, Y_grid.T, color='gray', linewidth=0.5)
axs[0].set_title("Grilla en coordenadas cartesianas")
axs[0].set_xlim(-6, 6)
axs[0].set_ylim(-6, 6)
axs[0].grid(True)

# Subplot 2: Grilla en coordenadas polares transformada a cartesianas
for i in range(R.shape[0]):
    axs[1].plot(X[i, :], Y[i, :], color='blue', linewidth=0.5)
for j in range(R.shape[1]):
    axs[1].plot(X[:, j], Y[:, j], color='red', linewidth=0.5)

axs[1].set_title("Grilla polar (en coordenadas cartesianas)")
axs[1].set_xlim(-6, 6)
axs[1].set_ylim(-6, 6)
axs[1].grid(True)

plt.suptitle("Transformacin del espacio: coordenadas cartesianas vs. polares", fontsize=14)
plt.tight_layout()
plt.show()
```
